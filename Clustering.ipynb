# Theoretical Questions:

1. What is unsupervised learning in the context of machine learning?

- Unsupervised learning is a type of machine learning where the model is trained on unlabeled data. The goal is to discover patterns, relationships, or groupings within the data without prior knowledge of the output.


2. How does K-Means clustering algorithm work?

-K-Means is a type of unsupervised learning algorithm that groups similar data points into K clusters. It works by:
- Initializing K centroids randomly.
- Assigning each data point to the nearest centroid.
- Updating the centroids based on the mean of the data points in each cluster.
- Repeating the process until convergence.


3. Explain the concept of a dendrogram in hierarchical clustering.

- A dendrogram is a tree-like structure used to visualize the hierarchical clustering process. It shows the merging or splitting of clusters at different levels of similarity, helping to identify the optimal number of clusters.


4. What is the main difference between K-Means and Hierarchical Clustering?

- The main difference is that K-Means requires specifying the number of clusters (K) beforehand, while Hierarchical Clustering builds a hierarchy of clusters without requiring a predefined number.


5. What are the advantages of DBSCAN over K-Means?

-DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages:
- Handles noise and outliers well.
- Does not require specifying the number of clusters.
- Can find clusters of varying shapes and sizes.


6. When would you use Silhouette Score in clustering?

- The Silhouette Score is used to evaluate the quality of clustering. It measures how similar a data point is to its own cluster compared to other clusters. It's useful for determining the optimal number of clusters.


7. What are the limitations of Hierarchical Clustering?

-Limitations include:
- Computationally expensive for large datasets.
- Sensitive to noise and outliers.
- Difficult to determine the optimal number of clusters.


8. Why is feature scaling important in clustering algorithms like K-Means?

- Feature scaling is important because K-Means uses distance-based calculations. If features have different scales, it can lead to biased clustering. Scaling ensures that all features contribute equally to the clustering process.


9. How does DBSCAN identify noise points?

- DBSCAN identifies noise points as those that do not belong to any cluster. A point is considered noise if it does not have enough neighbors within a specified radius (ε) to form a dense region.



10. Define inertia in the context of K-Means.

- Inertia in K-Means refers to the sum of squared distances between each data point and its assigned centroid. Lower inertia indicates better clustering, as data points are closer to their centroids. It's often used to determine the optimal number of clusters (K) by plotting inertia against different values of K.

11. What is the elbow method in K-Means clustering?
The elbow method is a technique used to determine the optimal number of clusters (K) in K-Means clustering. It involves plotting the sum of squared errors (inertia) against different values of K and selecting the K value where the rate of decrease in inertia becomes less steep, forming an "elbow" shape.



12. Describe the concept of "density" in DBSCAN.
In DBSCAN, density refers to the number of data points within a specified radius (ε) around a point. A point is considered a core point if it has a minimum number of points (MinPts) within this radius, indicating a dense region. Density-based clustering helps identify clusters of varying shapes and sizes.



13. Can hierarchical clustering be used on categorical data?
Yes, hierarchical clustering can be used on categorical data, but it requires a suitable distance metric that can handle categorical variables. Some common metrics include Hamming distance or Gower's distance. Additionally, the linkage criteria need to be chosen appropriately.



14. What does a negative Silhouette Score indicate?
A negative Silhouette Score indicates that a data point has been assigned to the wrong cluster. It means the point is closer to points in another cluster than to points in its own cluster, suggesting poor clustering.



15. Explain the term "linkage criteria" in hierarchical clustering.
Linkage criteria in hierarchical clustering refer to the method used to calculate the distance between clusters. Common linkage criteria include:
- Single linkage: Minimum distance between points in two clusters.
- Complete linkage: Maximum distance between points in two clusters.
- Average linkage: Average distance between points in two clusters.
- Ward's linkage: Minimizes the variance within clusters.



16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
K-Means assumes clusters are spherical and of similar size and density. When clusters have varying sizes or densities, K-Means may struggle to identify them accurately, leading to poor clustering performance.



17. What are the core parameters in DBSCAN, and how do they influence clustering?
The core parameters in DBSCAN are:
- ε (epsilon): The maximum distance between points in a cluster. A larger ε can merge clusters, while a smaller ε can create more clusters.
- MinPts: The minimum number of points required to form a dense region. A higher MinPts can reduce noise but may miss smaller clusters.



18. How does K-Means++ improve upon standard K-Means initialization?
K-Means++ improves initialization by spreading out the initial centroids. Instead of randomly selecting centroids, K-Means++ chooses the first centroid randomly and then selects subsequent centroids based on a probability distribution that favors points farther from existing centroids. This leads to better clustering performance and faster convergence.



19. What is agglomerative clustering?
Agglomerative clustering is a type of hierarchical clustering that starts with each data point as its own cluster and iteratively merges the closest clusters until all points are in a single cluster. It builds a hierarchy of clusters in a bottom-up manner.



20. What makes Silhouette Score a better metric than just inertia for model evaluation?
The Silhouette Score is a better metric than inertia because it provides a more interpretable measure of clustering quality. It evaluates how well each point fits into its cluster relative to other clusters, whereas inertia only measures the sum of squared distances within clusters. Silhouette Score helps assess the separation and cohesion of clusters more effectively.

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=4, cluster_std=0.8, random_state=0)
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.5)
plt.show()

```

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=4, cluster_std=0.8, random_state=0)
silhouette_scores = []

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

plt.bar(range(2, 6), silhouette_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, y = make_blobs(n_samples=200, centers=4, cluster_std=1.5, random_state=0)
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.show()
```
